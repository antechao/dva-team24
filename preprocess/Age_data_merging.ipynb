```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv(r'/Users/vivianchen/Desktop/CSE6242/project/webpage/BX_Books.csv', sep=';')
id_user = pd.read_csv(r'/Users/vivianchen/Desktop/CSE6242/project/webpage/BX-Book-Ratings.csv', sep=';')
user_age = pd.read_csv(r'/Users/vivianchen/Desktop/CSE6242/project/webpage/BX-Users.csv', sep=';')
```

```python
id_user = id_user[['User-ID', 'ISBN']]
id_user
```

```python
user_age = user_age[['User-ID', 'Age']]
user_age
```

```python
book = book[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-M']]
book.head()
```

```python
tmp = pd.merge(id_user, user_age, on='User-ID', how='left')
tmp
```

```python
final = pd.merge(tmp, book, on='ISBN')
final
```

```python
final.to_csv(r'/Users/vivianchen/Desktop/CSE6242/project/webpage/merged_data_m.csv', sep=',', index=False)
```
